```python
string1 = "abc"
string2 = "123"

# Initialize an empty string to store the merged result
merged_string = ""

# Iterate over both strings simultaneously
for char1, char2 in zip(string1, string2):
    # Append characters alternatively
    merged_string += char1 + char2

# If one string is longer than the other, append the remaining characters
if len(string1) > len(string2):
    merged_string += string1[len(string2):]
elif len(string2) > len(string1):
    merged_string += string2[len(string1):]

print(merged_string)
```

```python
def find_nth_occurrence(lst, element, n):
    try:
        occurrence_index = -1
        for _ in range(n):
            occurrence_index = lst.index(element, occurrence_index + 1)
        return occurrence_index
    except ValueError:
        return None

# Example usage:
a = [1, 2, 3, 1, 4, 1]
element = 1
n = 2
nth_index = find_nth_occurrence(a, element, n)
if nth_index is not None:
    print(f"The index of the {n}th occurrence of {element} is {nth_index}.")
else:
    print(f"There is no {n}th occurrence of {element} in the list.")
```

```python
map() funcation:
The map() function executes a specified function for each item in an iterable. The item is sent to the function as a parameter.

Syntax:
map(function, iterables)
```

```python

numbers = [1, 2, 3, 4, 5]

# Define a function to square a number
def square(x):
    return x ** 2

# Apply the square function to each number in the list using map
squared_numbers = map(square, numbers)

# Convert the map object to a list to see the results
squared_numbers_list = list(squared_numbers)

print(squared_numbers_list)  # Output: [1, 4, 9, 16, 25]
```

```python
def myfunc(a):
  return len(a)

x = map(myfunc, ('apple', 'banana', 'cherry'))

print(x)

#convert the map into a list, for readability:
print(list(x))
```

```python
import pickle

class Employee:
    def __init__(self, eno, ename, esal, eaddr):
        self.eno = eno
        self.ename = ename
        self.esal = esal
        self.eaddr = eaddr

    def display(self):
        print(self.eno, "\t", self.ename, "\t", self.esal, "\t", self.eaddr)

# Pickling the Employee object
with open("emp.dat", "wb") as f:
    e = Employee(100, "Durga", 1000, "Hyd")
    pickle.dump(e, f)
    print("Pickling of Employee Object completed...")

# Unpickling the Employee object
with open("emp.dat", "rb") as f:
    obj = pickle.load(f)
    print("Printing Employee Information after unpickling:")
    obj.display()
```

```python
class MyClass:
    def __init__(self, x):
        self.x = x
    
    # Instance Method
    def instance_method(self):
        return f"Instance method called with value {self.x}"
    
    # Class Method
    @classmethod
    def class_method(cls, y):
        return f"Class method called with value {y}"
    
    # Static Method
    @staticmethod
    def static_method(z):
        return f"Static method called with value {z}"

# Creating an instance of MyClass
obj = MyClass(5)

# Calling Instance Method
print(obj.instance_method())  # Output: Instance method called with value 5

# Calling Class Method
print(MyClass.class_method(10))  # Output: Class method called with value 10

# Calling Static Method
print(MyClass.static_method(15))  # Output: Static method called with value 15
```

# Python Inheritance


```python
class Animal:

    # attribute and method of the parent class
    name = ""
    
    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):

    # new method in subclass
    def display(self):
        # access name attribute of superclass using self
        print("My name is ", self.name)

# create an object of the subclass
labrador = Dog()

# access superclass attribute and method 
labrador.name = "Rohu"
labrador.eat()

# call subclass method 
labrador.display()
```

```python
class Animal:

    # attributes and method of the parent class
    name = ""
    
    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):

    # override eat() method
    def eat(self):
        print("I like to eat bones")

# create an object of the subclass
labrador = Dog()

# call the eat() method on the labrador object
labrador.eat()
```

```python
class Animal:

    name = ""
    
    def eat(self):
        print("I can eat")

# inherit from Animal
class Dog(Animal):
    
    # override eat() method
    def eat(self):
        
        # call the eat() method of the superclass using super()
        super().eat()
        
        print("I like to eat bones")

# create an object of the subclass
labrador = Dog()

labrador.eat()
```

```python
class Mammal:
    def mammal_info(self):
        print("Mammals can give direct birth.")

class WingedAnimal:
    def winged_animal_info(self):
        print("Winged animals can flap.")

class Bat(Mammal, WingedAnimal):
    pass

# create an object of Bat class
b1 = Bat()

b1.mammal_info()
b1.winged_animal_info()
```

```python
class SuperClass:

    def super_method(self):
        print("Super Class method called")

# define class that derive from SuperClass
class DerivedClass1(SuperClass):
    def derived1_method(self):
        print("Derived class 1 method called")

# define class that derive from DerivedClass1
class DerivedClass2(DerivedClass1):

    def derived2_method(self):
        print("Derived class 2 method called")

# create an object of DerivedClass2
d2 = DerivedClass2()

d2.super_method()  # Output: "Super Class method called"

d2.derived1_method()  # Output: "Derived class 1 method called"

d2.derived2_method()  # Output: "Derived class 2 method called"
```

```python
class SuperClass1:
    def info(self):
        print("Super Class 1 method called")

class SuperClass2:
    def info(self):
        print("Super Class 2 method called")

class Derived(SuperClass1, SuperClass2):
    pass

d1 = Derived()
d1.info()  

# Output: "Super Class 1 method called"
```

```python
class Cat:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def info(self):
        print(f"I am a cat. My name is {self.name}. I am {self.age} years old.")

    def make_sound(self):
        print("Meow")


class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def info(self):
        print(f"I am a dog. My name is {self.name}. I am {self.age} years old.")

    def make_sound(self):
        print("Bark")


cat1 = Cat("Kitty", 2.5)
dog1 = Dog("Fluffy", 4)

for animal in (cat1, dog1):
    animal.make_sound()
    animal.info()
    animal.make_sound()
```

```python
# Nested function in Python:

def greet(name):
    # inner function
    def display_name():
        print("Hi", name)
    
    # call inner function
    display_name()

# call outer function
greet("John")  

# Output: Hi John
```

```python
#Python Closures As we have already discussed, 
#closure is a nested function that helps us access the outer function's variables even after the outer function is closed. 
```

```python
def greet():
    # variable defined outside the inner function
    name = "John"
    
    # return a nested anonymous function
    return lambda: "Hi " + name

# call the outer function
message = greet()

# call the inner function
print(message())

# Output: Hi John
```

```python
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function

closure_func = outer_function(5)
result = closure_func(3)
print(result)  # Output will be 8
```

```python
import time

def countdown(num):
    print('countdown started ....')
    while num > 0:
        yield num
        num=num-1

g=countdown(6)

for i in g:
    print(i)
```

# Decorators

```python
def decor(func):
    def inner():
        func()   # existing functionality like hi am
        # add new functionality
        print("bhushan")
    return inner
    

def printer():  # i want to update it i want it like hi iam bhushan without update it
    print("hi")
    print("i am ")


printer=decor(printer)  # pass printer function as input to decor
printer()    # call original printer
    
```

```python
# 2nd method

def decor(printer):
    def inner():
        printer()   # existing functionality like hi am
        # add new functionality
        print("bhushan")
    return inner
    
@decor
def printer():  # i want to update it i want it like hi iam bhushan without update it
    print("hi")
    print("i am ")



printer()    # call original printer
    
```

```python
#  example 2

def decor(addition): # new update code
    def inner():
        result=addition()  # existiing functionality
        # add new functionality
        num3=float(input("enter the 3rd number:"))
        result = result+num3
        return result
    return inner
    
@decor
def addition():  # old code
    num1=float(input("enter the 1st number:"))
    num2=float(input("enter the 2nd number:"))
    result= num1+num2
    return result

addition()
# now new requirement is it has to ask user to add three numbers rather than 2
```
